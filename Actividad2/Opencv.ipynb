#Mi primer proyecto de vision computacional
Este primer proyecto es un ejemplo de como utilizar OpenCV de manera efectiva

```python
#Importar librerias
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
bgr = cv2.imread('Imagenpoo.webp')
print("El tama√±o de la imagen es: ", bgr.shape)

#Convertir a RGB
rgb = cv2.cvtColor(bgr,cv2.COLOR_BGR2RGB)

#Mostrar la imagen
plt.imshow(rgb)
```

```python
#Mostrar la imagen en escala de grises
gray = cv2.cvtColor(bgr,cv2.COLOR_BGR2GRAY)
plt.imshow(gray, cmap = 'gray')
```

```python
#Mostrar los componentes de color
plt.figure(figsize=(10,7))
plt.subplot(1,3,1)
plt.imshow(bgr[:,:,0], cmap='gray')
plt.title('Canal Azul')
plt.subplot(1,3,2)
plt.imshow(bgr[:,:,1], cmap='gray')
plt.title('Canal Verde')
plt.subplot(1,3,3)
plt.imshow(bgr[:,:,2], cmap='gray')
plt.title('Canal Rojo')
plt.show()
```

```python
#Moldear las componentes HLS de la imagen
hls = cv2.cvtColor(bgr,cv2.COLOR_BGR2HLS)
plt.figure(figsize=(10,7))
plt.subplot(1,3,1)
plt.imshow(hls[:,:,0], cmap='gray')
plt.title('Canal H')
plt.subplot(1,3,2)
plt.imshow(hls[:,:,1], cmap='gray')
plt.title('Canal L')
plt.subplot(1,3,3)
plt.imshow(hls[:,:,2], cmap='gray')
plt.title('Canal S')
plt.show()
```

```python
#Ajustar brillo y contraste
alpha = 1.5 # Contraste
beta = 50   # Brillo
adjusted = cv2.convertScaleAbs(bgr, alpha=alpha, beta=beta)
#Mostrar la imagen original y la ajustada
plt.figure(figsize=(10,7))
plt.subplot(1,2,1)
plt.imshow(cv2.cvtColor(bgr, cv2.COLOR_BGR2RGB))
plt.title('Imagen Original')
plt.subplot(1,2,2)
plt.imshow(cv2.cvtColor(adjusted, cv2.COLOR_BGR2RGB))
plt.title('Imagen con Brillo y Contraste Ajustados')
plt.show()
```

```python
#Aplicar un filtro Gaussiano
blurred = cv2.GaussianBlur(bgr, (15, 15), 0)
#Mostrar la imagen original y filtrada
plt.figure(figsize=(10,7))
plt.subplot(1,2,1)
plt.imshow(cv2.cvtColor(bgr, cv2.COLOR_BGR2RGB))
plt.title('Imagen Original')
plt.subplot(1,2,2)
plt.imshow(cv2.cvtColor(blurred, cv2.COLOR_BGR2RGB))
plt.title('Imagen con Filtro Gaussiano')
plt.show()
```
